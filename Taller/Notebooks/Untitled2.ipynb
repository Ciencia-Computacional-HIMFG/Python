```python
import numpy as np
from matplotlib import pyplot as plt
from tqdm import tqdm
```

```python
n_random = 10000000
random_pairs = np.random.uniform(low=-1.0, high=1.0, size=(n_random,2))
```

```python
contador = 0
contador2 = 0
dentro = 0
list_intentos = []
list_pis = []

for x,y in tqdm(random_pairs[:]):
    contador += 1
    if contador > 9000000:
        contador2 += 1
        if x**2+y**2 < 1.0:
            dentro +=  1
        list_intentos.append(contador)
        list_pis.append(4.0*dentro/contador2)
```

```python
plt.plot(list_intentos,list_pis)
plt.ylim([3.12, 3.16])
plt.hlines(3.141592, 0, n_random, colors='red')
plt.show()
```

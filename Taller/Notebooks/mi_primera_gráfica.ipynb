## Objetivos

Hacer una representación gráfica sencilla con ayuda de la librería Matplotlib.

## Resolución

Importamos en primer lugar el objeto `pyplot` de la librería `matplotlib` con el alias `plt`.

```python
from matplotlib import pyplot as plt
```

Definimos dos variables, `x` y `y`, con dos listas de números enteros.
Nuestra representación gráfica vendrá dada por los puntos cuya coordenadas X y Y son (0,3), (1,4), (2,4) y (3,5).

```python
x = [0,1,2,3]
y = [3,4,4,5]
```

Invocamos la función `plot` del módulo `pyplot` de la librería Matplotlib que hemos importado con el alias `plt`.

```python
plt.plot(x,y)
```

# Cálculo del número Pi mediante método MonteCarlo

```python
import numpy as np
from matplotlib import pyplot as plt
from tqdm import tqdm
```

```python
n_random = 10000000
random_pairs = np.random.uniform(low=-1.0, high=1.0, size=(n_random,2))
```

```python
contador = 0
dentro = 0
list_intentos = []
list_pis = []

for x,y in tqdm(random_pairs[:]):
    contador += 1
    if x**2+y**2 < 1.0:
        dentro +=  1
    list_intentos.append(contador)
    list_pis.append(4.0*dentro/contador)
```

```python
print(f"El valor de pi estimado es {list_pis[-1]}")
```

```python
plt.plot(list_intentos,list_pis)
plt.ylim([3.1, 3.2])
plt.hlines(3.141592, 0, n_random, colors='red')
plt.title("Pi estimado dependiente de la cantidad de pares aleatorios")
plt.xlabel("Cantidad de pares aleatorios")
plt.ylabel("Pi estimado")
plt.show()
```
